Understand Missing Data

```python
import pandas as pd 
df = pd.read_csv("/kaggle/input/datasets/yasserh/titanic-dataset/Titanic-Dataset.csv")
 df.isnull().sum()  #df.isnull() → converts your dataset into True/False

# True = missing value
# False = not missing
# Meaning:
# ✅ No missing values:
# PassengerId, Survived, Pclass, Name, Sex, SibSp, Parch, Ticket, Fare
# ⚠️ Missing values:
# Age → 177 missing
# Cabin → 687 missing (very high!)
# Embarked → 2 missing
```

```python
df = pd.read_csv("/kaggle/input/datasets/yasserh/titanic-dataset/Titanic-Dataset.csv")
df.fillna(df.mean(numeric_only=True)) #Calculates mean of only numeric columns like Age,fare
#Age → mean value (e.g., ~29.7
# 888  ...  Age = 29.699118
# That value is not a real age
# 👉 It is the mean age that got filled in
```
