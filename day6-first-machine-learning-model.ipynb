*Step 1*
**Understand the Task:-**
**We’ll use:
Titanic dataset**

**👉 Goal:
Predict → passenger survived or no**

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/datasets/yasserh/titanic-dataset/Titanic-Dataset.csv")
df["Sex"].unique()  #survived (1) or did not survive (0)
```

**Step 2: Prepare Data**

```python
##Select important columns
df = df[["Survived", "Pclass", "Sex", "Age"]]
# the "Pclass", "Sex", "Age" are the inputs for the target survived or not 

##Convert text → numbers: or Converted categorical data to numerical
df["Sex"] = df["Sex"].astype(str).str.lower().str.strip()
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
## Why?
# ML models cannot understand text
# So:
# male → 0
# female → 1

## Handle missing values:
df = df.fillna(df.mean(numeric_only=True))
# Why?
# Some Age values are missing 😕
# 👉 ML models cannot handle empty values

# So we fill with average (mean)
df.head()
```

**Step 3: Split Data**

```python
from sklearn.model_selection import train_test_split

X = df[["Pclass", "Sex", "Age"]]  # Inputs
y = df["Survived"]               # Output

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
#You divide data into:
# Training data (80%) → model learns from this
# Testing data (20%) → model is tested on this
# 👉 Why?
# Train model on some data
# Test on unseen data
X_train.shape, X_test.shape
#  X_train.head() , is used to see the training data , and it should hve Pcls,sex,Age
#y_train.head() # is used to check the target
```

👉 Think like this:

X = what model uses to learn,
y = what model tries to predict
this:- train_test_split(...)  , is used to split the data
. What is test_size=0.2?
👉 Means:
80% → Training
20% → Testing

**Step 4: Train Model**

```python
from sklearn.linear_model import LogisticRegression
#👉 You are importing a machine learning algorithm
# 📌 What is Logistic Regression?
# 👉 A model used for classification problems

model = LogisticRegression()## creating a model
#You are creating an empty brain 🧠
# At this point:
# ❌ It knows nothing
# ❌ No learning yet

model.fit(X_train, y_train)
# 🧠 What does .fit() do?
# 👉 It tells the model:
# “Look at this data and learn patterns
```

**Step 5: Predict**

```python
predictions = model.predict(X_test)
# 🧠 What is happening?
# 👉 Your trained model is now making predictions on new data
# Remember:
# It has never seen X_test before
# So this is a real test

#See first few predictions:
predictions[:10]
```

**🚀 🔹 STEP 6: Check Accuracy**

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)
print(accuracy)
# 🧠 What is happening?

# 👉 You are comparing:
# Actual answers → y_test
# Model answers → predictions
```
