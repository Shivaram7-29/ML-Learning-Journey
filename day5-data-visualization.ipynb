**Basic Graph**

```python
import matplotlib.pyplot as plt
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y) #if u just write untill here u get this [<matplotlib.lines.Line2D at 0x7ca058271be0>]
# to avoid that in output we call
plt.title('Our first Graph!')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
## u can also play with the font
import matplotlib.pyplot as plt
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.title('Our first Graph!', fontdict={'fontname':'serif','fontsize':15})
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
## U can also change those ticks like values in graph 
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)
plt.title('Our first Graph!', fontdict={'fontname':'serif','fontsize':15})
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.show()
```

**Bar Chart**

```python
import matplotlib.pyplot as plt
labels = ['A','B','C']
values = [1,4,2]

bars=plt.bar(labels,values)
bars[0].set_hatch('/') #by using this line u see in o/p the pattern 
plt.show()
```

**Real world**

```python
#Step 1: Setup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/kaggle/input/datasets/shivram7/gas-price/gas_prices.csv")
df
```

```python
print(df.columns)
```

```python
#Step 2: Histogram
df["USA"].hist()
plt.xlabel('prices')
plt.show()
#Most values are on the left side
# Around 1.0 – 1.5
# ➡️ Meaning:
# Most gas prices were low
# Very few values on the right side
#  Around 2.5 – 3.2
# ➡️ Meaning:
# High gas prices happened rarely

```

```python
#LINE PLOT
plt.plot(df["Year"], df["USA"])
plt.xlabel('Year')
plt.show() #Gas prices increased after 2005
```

```python
tit["Sex"].value_counts().plot(kind="bar")
plt.show()
```

```python
sns.boxplot(x=tit["Age"])
plt.show()
#Where most data lies
# How spread out it is
# If there are outliers (extreme values)
```
