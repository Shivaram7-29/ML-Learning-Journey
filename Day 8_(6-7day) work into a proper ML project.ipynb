```python
#STEP 1
import pandas as pd 
df = pd.read_csv("/kaggle/input/datasets/yasserh/titanic-dataset/Titanic-Dataset.csv")
#Clean final dataset
df = df[["Survived", "Pclass", "Sex", "Age", "SibSp", "Fare"]]
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df = df.fillna(df.mean(numeric_only=True))
#👉 Keep only useful features → better model + cleaner project
```

**Step 2: Proper Train-Test Split**

```python
from sklearn.model_selection import train_test_split

X = df.drop("Survived", axis=1)
# 👉 All columns except target(survived)
# Example:
# Pclass
# Sex
# Age
# SibSp
# Fare
y = df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```


****Step 3: Train BEST Model
👉 Use Decision Tree (from Day 7)****


```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

**Step 4: Evaluate Properly**

```python
from sklearn.metrics import accuracy_score
pred = model.predict(X_test)
print(accuracy_score(y_test,pred))  #y_test:-- original answers and pred:-- predicted or u worte answers in exam
```

**Step 5: Add Basic Visualization**

```python
import matplotlib.pyplot as plt
df["Pclass"].value_counts().plot(kind="bar")
plt.xticks([0, 1, 2], ["3rd Class", "1st Class", "2nd Class"])
plt.show()
```
