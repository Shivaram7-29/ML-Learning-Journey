```python
#Let's deal with series:---
# series = A Pandas 1-Dimensional labeled array that  can hold any data type,think of it like a single
# column in a spreadsheet
import pandas as pd
data = [100,102,104] #or data = ["a","b","c"] this dtype: object

series = pd.Series(data)  # Now, this Series(), is a constructor with arguments
print(series)
```

```python
# WE can customize the index 
data = [100,102,104] 

series = pd.Series(data,index=["a","b","c"])  
print(series)
```

```python
# IF u want to access a value directly in series , u can do it with .loc
data = [100,102,104] 

series = pd.Series(data,index=["a","b","c"])  
print(series.loc["a"])   # i,e. return the value where the label is a
```

```python
# U can also locate a value i.e, u can change a value like:---
data = [100,102,104] 

series = pd.Series(data,index=["a","b","c"])  
series.loc["a"]=200
print(series)
```

```python
# with .iloc , but .iloc only deals with int index like the original one 0,1,....
data = [100,102,104] 

series = pd.Series(data,index=["a","b","c"])  
print(series.iloc[0])
```

```python
data = [100,102,104,205,250] 

series = pd.Series(data,index=["a","b","c","d","e"])  
print(series[series>200])
```

DATA FRAME

```python
# DATAFRAME = A tabular data structure with rows& columns(2 DIMENSIONAL)
import pandas as pd

data= {
    "Name": ["mia","rose", "jack"],
    "Age": [20, 30, 50]
}
# Now we will convert this dic to a dataframe
df = pd.DataFrame(data) #Now this DataFrame is a constructor 
print(df)   # now the o/p is a tabular one 
```

```python
data= {
    "Name": ["mia","rose", "jack"],
    "Age": [20, 30, 50]
}
df = pd.DataFrame(data, index=["Emp1","Emp2","Emp3"]) 
print(df)
print(df.loc["Emp1"])  # loc means location by label 
```

```python
data= {
    "Name": ["mia","rose", "jack"],
    "Age": [20, 30, 50]
}
df = pd.DataFrame(data, index=["Emp1","Emp2","Emp3"]) 

# ADD A NEW COLUMN
df["Job"]=["Dancer","N/A","CRICKETER"]

#ADD a new ROW
new_row = pd.DataFrame([{"Name":"Sandy","Age":27,"Job":"actress"}],index=["Emp4"])
#Now we need to concantinate this to the original one
df=pd.concat([df,new_row])
print(df)
```
